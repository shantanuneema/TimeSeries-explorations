```python
import pandas as pd
import numpy as np
import matplotlib as mpl

mpl.style.use('ggplot')
%matplotlib inline

df_air = pd.read_csv('../Data/airline_passengers.csv', index_col = 'Month', parse_dates = True)
df_air.dropna(inplace = True)
print(df_air.head())
df_air.plot();
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(df_air['Thousands of Passengers'], model='multiplicative')  # model='mul' also works
result.plot();
```
